```python
def solid_read(file):
    return pd.read_table(file,delimiter=' ',index_col=0,names=['x','y','z','sxb','syb','szb','sxk','syk','szk','temp']).dropna(axis=1)

def liquid_read(file):
    return pd.read_table(file,delimiter=' ',index_col=0,names=['x','y','z','sxp','syp','szp','sxb','syb','szb','sxk','syk','szk','temp']).dropna(axis=1).reset_index()
```

```python
#Read files -- reference and composite 

# K and epsilon values to be read
K_values=np.array(['5','10','107'])
e_values=np.array(['05','14','22'])
refs=np.array([]);
systems=np.array([]);
for k in K_values:
    str1="ref/stress.{}.out"
    f=str1.format(k)
    var1="ref_{}"
    var1=var1.format(k);
    refs=np.append(refs,var1)
    exec(var1 + "=solid_read(f)")
    for e in e_values:
        str2="composite/stress.{}.{}.out"
        ff=str2.format(k,e)
        var2="lq_{}_{}"
        var2=var2.format(k,e)
        exec(var2 + "=liquid_read(ff)")
        systems=np.append(systems,var2)

#Define Layers
for i in range(11):
    temp="l{}"
    temp=temp.format(i)
    exec(temp+"=np.arange((i-1)*3600,i*3600)")
```

```python
## Define basic data variables -- assumes a specific system size 
ny=600;nx=3750;
xgrid=np.linspace(0,375,nx);ygrid=np.linspace(0,60,ny)
Xgrid,Ygrid=np.meshgrid(xgrid,ygrid)

## define a function to read data and interpolate to plot contours
def stress_analysis(what,interactions):
    atoms=[what['x'],what['y']]
    points=pd.concat(atoms,axis=1)
    R=np.vstack((Xgrid.reshape(nx*ny),Ygrid.reshape(nx*ny))).T
    method='cubic';
    if interactions=='bond':
        s1='sxb';s2='syb';s3='szb';
    elif interactions=='pair':
        s1='sxp';s2='syp';s3='szp';
    else:
        s1='sxk';s2='syk';s3='szk';
        
    z1=what[s1];
    fmap1=interpolate.griddata(points,z1,R,method=method);
    z2=what[s2];
    fmap2=interpolate.griddata(points,z2,R,method=method);
    z3=what[s3];
    fmap3=interpolate.griddata(points,z3,R,method=method);
    return fmap1,fmap2,fmap3
```
